```python
import pandas as pd

df = pd.read_csv("employment-rates.csv")
df.rename(columns={'Pacific Island Countries and territories':'island'}, inplace=True)
print(df.columns)
df = df[df["INDICATOR"]=="UNEMPRATE"]
df = df[df["Sex"].isin(["Male", "Female"])]
df = df[df["Age"].isin(["25-54", "55-64"])]
df = df[["island", "TIME_PERIOD", "OBS_VALUE", "Age", "Sex"]]
df.columns = ["island", "year", "unemployment_rate", "age", "sex"]
df.sort_values(by=["island", "year", "age", "sex"])
```
