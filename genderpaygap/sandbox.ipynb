```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("genderpaygap.csv").dropna(axis=1)
df = df.loc[:, df.nunique() > 1]
df.drop(columns=['URBANIZATION', 'OCCUPATION'], inplace=True)
df = df.sort_values(by=["TIME_PERIOD", "Pacific Island Countries and territories", "Urbanization"])
df.drop(columns=["GEO_PICT"])
#df.to_csv("genderPayGap_perYear_perLocation_perUrbanization.csv", index=False)
df
```

```python
pivot_df = df.pivot_table(index='TIME_PERIOD', columns='Urbanization', values='OBS_VALUE', aggfunc='mean').fillna(0)

# Plot
pivot_df.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Time Period')
plt.ylabel('Observational Value')
plt.title('Grouped Bar Chart of Gender Pay Gap Over Time')
plt.legend(title='Occupation')
plt.grid(True)
plt.show()
```
